```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Resources/spotify.csv")
df.tail(5)
```

```python
df.describe()
```

```python
popular_artists_df = df[df['artist_popularity'] >= 65]
```

```python
popular_artists_df.describe()
```

```python
pop_tracks_pop_artists = popular_artists_df[popular_artists_df['track_popularity'] >= 52]
```

```python
x = pop_tracks_pop_artists.track_popularity

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3,3, figsize=(15, 15))
ax1.scatter(x, pop_tracks_pop_artists['acousticness'], s=1)
ax2.scatter(x, pop_tracks_pop_artists['danceability'], s=1)
ax3.scatter(x, pop_tracks_pop_artists['energy'], s=1)
ax4.scatter(x, pop_tracks_pop_artists['instrumentalness'], s=1)
ax5.scatter(x, pop_tracks_pop_artists['valence'], s=1)
ax6.scatter(x, pop_tracks_pop_artists['liveness'], s=1)
ax7.scatter(x, pop_tracks_pop_artists['speechiness'], s=1)
ax8.scatter(x, pop_tracks_pop_artists['tempo'], s=1)
ax9.scatter(x, pop_tracks_pop_artists['loudness'], s=1)

ax1.set_title('Popular Acousticness')
ax2.set_title('Popular Danceability')
ax3.set_title('Popular Energy')
ax4.set_title('Popular Instrumentalness')
ax5.set_title('Popular Valence')
ax6.set_title('Popular Liveness')
ax7.set_title('Popular Speechiness')
ax8.set_title('Popular Tempo')
ax9.set_title('Popular Loudness')

plt.savefig('pop_tracks_pop_artists.png')
```
